## What Are The Most Popular Paint Colors?

```python
df['paint_color'].value_counts().plot.area();
```

## What Is The Most Popular Model?

```python
sns.countplot(y=df['model'],order=df['model'].value_counts().index[2:7])
plt.title('Top 5 Models')
plt.show()
```

## What Is The Most Popular Region?

```python
sns.countplot(y=df['region'],order=df['region'].value_counts().index[2:7])
plt.title('Top 5 Regions')
plt.show()
```

## What Is The Most Pupular Type?

```python
sns.countplot(x='type',data=df,order=df['type'].value_counts().index[2:7]);
plt.show();
```
